```python
KfoldCV = StratifiedKFold(shuffle = True)

scaler = MinMaxScaler()
transformer = PowerTransformer()

HGBCParams = {"learning_rate" : (0.001, 0.5),
             "max_iter" : (100, 1000),
             "max_depth" : (4, 20),
             "l2_regularization" : (0, .5),
             "max_leaf_nodes" : (50, 500),
             "min_samples_leaf" : (10, 50)}

def HGBCModel(learning_rate, max_iter, max_depth, l2_regularization, max_leaf_nodes, min_samples_leaf) :
    max_iter = int(max_iter)
    max_depth = int(max_depth)
    max_leaf_nodes = int(max_leaf_nodes)
    min_samples_leaf = int(min_samples_leaf)
    
    HGBC_acc = np.empty(shape = (0, ))

    for train_id, valid_id in KfoldCV.split(trainFeature, trainTarget) :
        train, valid = trainFeature.iloc[train_id, :], trainFeature.iloc[valid_id, :]

        train = pd.DataFrame(scaler.fit_transform(train), columns = trainFeature.columns)
        valid = pd.DataFrame(scaler.transform(valid), columns = trainFeature.columns)

        train = pd.DataFrame(transformer.fit_transform(train), columns = trainFeature.columns)
        valid = pd.DataFrame(transformer.transform(valid), columns = trainFeature.columns)

        trainlabel, validlabel = trainTarget.iloc[train_id], trainTarget.iloc[valid_id]

        HGBC = HistGradientBoostingClassifier(learning_rate = learning_rate, max_iter = max_iter, max_depth = max_depth,
                                              l2_regularization = l2_regularization, max_leaf_nodes = max_leaf_nodes,
                                              min_samples_leaf = min_samples_leaf)
        
        HGBC.fit(train, trainlabel)
        
        HGBC_acc = np.append(HGBC_acc, HGBC.score(valid, validlabel))
        
    return HGBC_acc.mean()

HGBCoptim = BayesianOptimization(f = HGBCModel, pbounds = HGBCParams, verbose = 1)
HGBCoptim.maximize(init_points = 10, n_iter = 100)
```

```python
import lightgbm as lgbm

KfoldCV = StratifiedKFold(shuffle = True)

scaler = MinMaxScaler()
transformer = PowerTransformer()

LGBMParams = {"num_leaves" : (50, 500),
              "n_estimators" : (100, 1000),
              "max_depth" : (3, 20),
              "min_split_gain" : (0, 0.1),
              "min_child_weight" : (0.001, 0.1),
              "min_child_samples" : (10, 50),
              "subsample" : (.5, 1.0),
              "subsample_freq" : (0, 5),
              "colsample_bytree" : (.5, 2),
              "reg_alpha" : (0, 5),
              "reg_lambda" : (0, 5),
              "bagging_freq" : (2, 10),
              "feature_fraction" : (.5, 1)}

def LGBMModel(num_leaves, n_estimators, max_depth, min_split_gain, min_child_weight, min_child_samples, subsample, subsample_freq, colsample_bytree, reg_alpha, reg_lambda, bagging_freq, feature_fraction) :
    num_leaves = int(num_leaves)
    max_depth = int(max_depth)
    n_estimators = int(n_estimators)
    min_child_samples = int(min_child_samples)
    bagging_freq = int(bagging_freq)
    
    LGBM_acc = np.empty(shape = (0, ))

    for train_id, valid_id in KfoldCV.split(trainFeature, trainTarget) :
        train, valid = trainFeature.iloc[train_id, :], trainFeature.iloc[valid_id, :]

        train = pd.DataFrame(scaler.fit_transform(train), columns = trainFeature.columns)
        valid = pd.DataFrame(scaler.transform(valid), columns = trainFeature.columns)

        train = pd.DataFrame(transformer.fit_transform(train), columns = trainFeature.columns)
        valid = pd.DataFrame(transformer.transform(valid), columns = trainFeature.columns)

        trainlabel, validlabel = trainTarget.iloc[train_id], trainTarget.iloc[valid_id]

        LGBM = lgbm.LGBMClassifier(num_leaves = num_leaves, n_estimators = n_estimators, max_depth = max_depth, min_split_gain = min_split_gain, min_child_weight = min_child_weight,
                                   min_child_samples = min_child_samples, subsample = subsample, subsample_freq = subsample_freq, colsample_bytree = colsample_bytree,
                                   reg_alpha = reg_alpha, reg_lambda = reg_lambda, bagging_freq = bagging_freq, feature_fraction = feature_fraction)
        
        LGBM.fit(train, trainlabel)
        
        LGBM_acc = np.append(LGBM_acc, LGBM.score(valid, validlabel))
        
    return LGBM_acc.mean()

LGBMoptim = BayesianOptimization(f = LGBMModel, pbounds = LGBMParams, verbose = 2)
LGBMoptim.maximize(init_points = 10, n_iter = 100)
```
